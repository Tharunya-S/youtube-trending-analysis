```python
data['category'].fillna(data['category'].mode()[0], inplace=True)
```

```python
data['Country'].fillna(data['Country'].mode()[0], inplace=True)
```

```python
data['Country'] = data['Country'].fillna(data['Country'].mode()[0])
data['Abbreviation'] = data['Abbreviation'].fillna(data['Abbreviation'].mode()[0])
data['channel_type'] = data['channel_type'].fillna(data['channel_type'].mode()[0])
data['created_month'] = data['created_month'].fillna(data['created_month'].mode()[0])
```

```python
data[['category', 'Country', 'Abbreviation', 'channel_type', 'created_month']].isnull().sum()
```

```python
data['country_rank'].fillna(data['country_rank'].median(), inplace=True)
```

```python
data['country_rank'] = data['country_rank'].fillna(data['country_rank'].median())
data['channel_type_rank'] = data['channel_type_rank'].fillna(data['channel_type_rank'].median())
data['video_views_for_the_last_30_days'] = data['video_views_for_the_last_30_days'].fillna(data['video_views_for_the_last_30_days'].median())
data['subscribers_for_last_30_days'] = data['subscribers_for_last_30_days'].fillna(data['subscribers_for_last_30_days'].median())
data['created_year'] = data['created_year'].fillna(data['created_year'].median())
data['created_date'] = data['created_date'].fillna(data['created_date'].median())
data['Gross tertiary education enrollment (%)'] = data['Gross tertiary education enrollment (%)'].fillna(data['Gross tertiary education enrollment (%)'].median())
data['Population'] = data['Population'].fillna(data['Population'].median())
data['Unemployment rate'] = data['Unemployment rate'].fillna(data['Unemployment rate'].median())
data['Urban_population'] = data['Urban_population'].fillna(data['Urban_population'].median())
data['Latitude'] = data['Latitude'].fillna(data['Latitude'].median())
data['Longitude'] = data['Longitude'].fillna(data['Longitude'].median())
```
