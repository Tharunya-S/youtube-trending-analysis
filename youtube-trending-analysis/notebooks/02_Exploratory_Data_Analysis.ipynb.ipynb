**Exploratory Data Analysis (EDA) â Univariate, Bivariate and
Multivariate Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Global YouTube Statistics.csv', encoding='latin1')
data
```

**Conduct Univariate Analysis to study the distribution and characteristics of individual
variables such as Subscribers, Views, Earnings, Country, and Category.**

### Univariate Analysis of Numerical Variables

```python
print("Descriptive statistics for 'subscribers':")
display(data['subscribers'].describe())

print("\nDescriptive statistics for 'video views':")
display(data['video views'].describe())
```

```python
plt.figure(figsize=(5, 3))
plt.subplot(1, 2, 1)
sns.histplot(data['subscribers'].dropna(), bins=30, kde=True)
plt.title('Distribution of Subscribers')
plt.xlabel('Subscribers')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(data['video views'].dropna(), bins=30, kde=True)
plt.title('Distribution of Video Views')
plt.xlabel('Video Views')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Univariate Analysis of Categorical Variables

```python
plt.figure(figsize=(5, 3))
sns.countplot(data=data, y='Country', order=data['Country'].value_counts().index[:10])
plt.title('Top 10 Countries by Youtuber Count')
plt.xlabel('Number of Youtubers')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=data, y='category', order=data['category'].value_counts().index[:10])
plt.title('Top 10 Categories by Youtuber Count')
plt.xlabel('Number of Youtubers')
plt.ylabel('Category')
plt.show()
```

### Bivariate Analysis

#### Subscribers vs Views

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(data=data, x='subscribers', y='video views', alpha=0.6)
plt.title('Subscribers vs Video Views')
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.xscale('log') # Use log scale due to large range of subscribers
plt.yscale('log') # Use log scale due to large range of video views
plt.grid(True, which="both", ls="--", c='0.7')
plt.show()
```

#### Category vs Average Views

```python
avg_views_by_category = data.groupby('category')['video views'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(5, 3))
sns.barplot(x=avg_views_by_category.values, y=avg_views_by_category.index, palette='viridis')
plt.title('Top 10 Categories by Average Video Views')
plt.xlabel('Average Video Views')
plt.ylabel('Category')
plt.show()
```

#### Earnings vs Subscribers

The DataFrame does not contain a direct 'Earnings' column. If you have a method to calculate or infer 'Earnings', please provide it, and I can proceed with this analysis.

```python
# Assuming an average CPM (Cost Per Mille) of $2 (per 1000 views) and a 50% monetization rate
# These are illustrative figures, actual earnings vary widely.
estimated_cpm = 2 # $2 per 1000 views
monetization_rate = 0.5 # 50% of views are monetized

data['estimated_earnings'] = (data['video views'] / 1000) * estimated_cpm * monetization_rate

print("Descriptive statistics for 'estimated_earnings':")
display(data['estimated_earnings'].describe())
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(data=data, x='subscribers', y='estimated_earnings', alpha=0.6)
plt.title('Subscribers vs Estimated Earnings')
plt.xlabel('Subscribers')
plt.ylabel('Estimated Earnings')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="--", c='0.7')
plt.show()
```

#### Country vs Number of Channels

This analysis was covered during the univariate analysis with the plot 'Top 10 Countries by Youtuber Count'.

```python
plt.figure(figsize=(5, 3))
sns.countplot(data=data, y='Country', order=data['Country'].value_counts().index[:10])
plt.title('Top 10 Countries by Youtuber Count')
plt.xlabel('Number of Channels')
plt.ylabel('Country')
plt.show()
```


## Multivariate Analysis



## Correlation Matrix of Numerical Variables



```python
numerical_cols = ['subscribers', 'video views', 'uploads', 'Gross tertiary education enrollment (%)',
                  'Population', 'Unemployment rate', 'Urban_population', 'estimated_earnings']

correlation_matrix = data[numerical_cols].corr()

plt.figure(figsize=(7,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Key Numerical Variables')
plt.show()
```

## Pairwise Relationships




```python
selected_cols_pairplot = ['subscribers', 'video views', 'uploads', 'estimated_earnings']

sns.pairplot(data[selected_cols_pairplot])
plt.suptitle('Pairwise Relationships of Key Numerical Variables', y=1.02) # Adjust y to prevent title overlap
plt.show()
```

```python
plt.figure(figsize=(7, 4))
channel_type_order = data.groupby('channel_type')['subscribers'].median().sort_values(ascending=False).index
sns.boxplot(data=data, x='subscribers', y='channel_type', order=channel_type_order, palette='viridis')
plt.title('Subscribers Distribution by Channel Type')
plt.xlabel('Subscribers')
plt.ylabel('Channel Type')
plt.xscale('log') # Use log scale due to large range of subscribers
plt.grid(True, which="both", ls="--", c='0.7')
plt.show()
```

```python
plt.figure(figsize=(7, 4))
channel_type_order = data.groupby('channel_type')['subscribers'].median().sort_values(ascending=False).index
sns.boxplot(data=data, x='subscribers', y='channel_type', order=channel_type_order, palette='viridis', hue='channel_type', legend=False)
plt.title('Subscribers Distribution by Channel Type')
plt.xlabel('Subscribers')
plt.ylabel('Channel Type')
plt.xscale('log') # Use log scale due to large range of subscribers
plt.grid(True, which="both", ls="--", c='0.7')
plt.show()
```

```python
total_views_by_country = data.groupby('Country')['video views'].sum().sort_values(ascending=False)
top_10_countries_by_views = total_views_by_country.head(10).index.tolist()

data_filtered_by_top_countries = data[data['Country'].isin(top_10_countries_by_views)]

plt.figure(figsize=(7, 4))
sns.boxplot(data=data_filtered_by_top_countries, x='video views', y='Country', order=data_filtered_by_top_countries.groupby('Country')['video views'].median().sort_values(ascending=False).index, palette='viridis', hue='Country', legend=False)
plt.title('Video Views Distribution by Top Countries')
plt.xlabel('Video Views')
plt.ylabel('Country')
plt.xscale('log')
plt.grid(True, which="both", ls="--", c='0.7')
plt.show()
```
